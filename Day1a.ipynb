write a program to keep asking for a number until you enter a negative number. At the end, print the sum of all entered numbers.


```python
user_input = float(input("please enter a negative number: "))
while user_input >= 0:
  print("Guy, your input has to be a gegative (-ve) number")
  user_input = float(input("please enter a negative number: "))

print(user_input)
```

```python
user_input = float(input("please enter a negative number: "))
all_positive_number = []
while user_input >= 0:
  all_positive_number.append(user_input)
  print("Guy, your input has to be a gegative (-ve) number")
  user_input = float(input("please enter a negative number: "))


total = sum(all_positive_number)
print(user_input)
print(f"Total positive number is: {total}")
```
